```python
import pandas as pd
from plotnine import *

data_file = "../data/render_info.csv"
data = pd.read_csv(data_file)

data.info()
```

```python
y_cols = list(data.columns)

del y_cols[0]
del y_cols[0]

y_cols
```

```python
x_label = "# of Points Plotted"
y_labels = [
    "Total Time (seconds)",
    "Max Memory Usage (GB)",
    "Average System CPU Usage (%)",
    "Time to Create Plot (seconds)",
    "Render Time (seconds)"
]

titles = [
    "Total Time v Points to Plot and Render",
    "Max Memory Usage v Points to Plot and Render",
    "Average CPU Usage v Points to Plot and Render",
    "Time to Plot v Points",
    "Time to Render v Points"
]
```

```python
for (col, y_lab, t) in zip(y_cols, y_labels, titles):
    print(
        ggplot(data, aes(x = "points", y = col)) +\
            facet_wrap("~type") +\
            geom_line() +\
            xlab(x_label) +\
            ylab(y_lab) +\
            ggtitle(t)
    )
```

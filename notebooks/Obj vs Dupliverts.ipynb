```python
import pandas as pd
from plotnine import *

cpu_obj_file = "../data/cpu_obj.txt"
cpu_obj = pd.read_csv(cpu_obj_file, header=None, names=["cpu"])
cpu_obj["time"] = range(len(cpu_obj))

cpu_dup_file = "../data/cpu_dupliverts.txt"
cpu_dup = pd.read_csv(cpu_dup_file, header=None, names=["cpu"])
cpu_dup["time"] = range(len(cpu_dup))
```

```python
ggplot(cpu_obj, aes(x = "time", y = "cpu")) +\
    geom_line() +\
    xlab("Interval") +\
    ylab("Total System CPU Usage (%)") +\
    ggtitle("Obj CPU usage on model drag")
```

```python
ggplot(cpu_dup, aes(x = "time", y = "cpu")) +\
    geom_line() +\
    xlab("Interval") +\
    ylab("Total System CPU Usage (%)") +\
    ggtitle("Dupliverts CPU usage on model drag")
```

```python
memory_file = "../data/obj_v_dupliverts_memory.csv"
memory = pd.read_csv(memory_file)
```

```python
ggplot(memory, aes(x = "type", y = "memory")) +\
    geom_col() +\
    xlab("Scatter Plot Type") +\
    ylab("Memory usage on model drag (GB)") +\
    ggtitle("Memory usage on drag by Scatter Plot Type")
```

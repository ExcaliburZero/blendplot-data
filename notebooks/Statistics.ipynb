```python
import pandas as pd
from plotnine import *
```

```python
data_file = "../data/time_memory_usage.csv"
data = pd.read_csv(data_file)

data.info()
```

```python
columns = list(data.columns)[1:]
column_descriptions = [
      "Time to generate obj file (sec)"
    , "Time to import obj file (sec)"
    , "Time to save blend file (sec)"
    , "Time to plot and save graph (sec)"
    , "Max Blender RAM usage (GB)"
    , "obj file size (MB)"
    , "blend file size (MB)"
    , "Total disk usage (MB)"
]
```

```python
for col, desc in zip(columns, column_descriptions):
    plot = ggplot(data, aes(x = "points", y = col)) +\
        geom_point() +\
        xlab("Points plotted") +\
        ylab(desc)
    print(plot)
```
